```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import random
```

```python
# Set random seed for reproducibility
np.random.seed(42)
feature_1 = np.random.uniform(-1, 1, 1000)
feature_2 = np.random.uniform(-1, 1, 1000)
feature_3 = np.random.uniform(-1, 1, 1000)

# Create a linear decision boundary
target = (feature_1 + feature_2 - feature_3 > 0).astype(int)

# Create a DataFrame
df = pd.DataFrame({
    'feature_1': feature_1,
    'feature_2': feature_2,
    'target': target
})
df.head()
```

```python
# That data crete by me for practics
df.info()
```

```python
df.describe()
```

```python
# Visualize the data
sns.scatterplot(x=df['feature_1'], y=df['feature_2'], hue=df['target'])
plt.title('Scatter plot of feature_1 vs feature_2')
```

```python
X=df[['feature_1', 'feature_2']]
y=df['target']
from sklearn.linear_model import Perceptron
pc=Perceptron(max_iter=1000, tol=1e-3, random_state=42)
pc.fit(X, y)
```

```python
pc.intercept_
```

```python
pc.coef_
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X.values, y.values, clf=pc, legend=2)
plt.xlabel('feature_1')
plt.ylabel('feature_2')
plt.title('Decision Boundary of Perceptron')
plt.show()
```
