```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Data.csv')
df.head()
```

```python
# The data has no missing values
# The data has no duplicates
# The data has no outliers
# The data has no categorical variables
# The data has no text variables
# The data has no date variables
# The data has no image variables
```

```python
sns.scatterplot(x=df['cgpa'], y=df['resume_score'], hue=df['placed'])
```

```python
X = df[['cgpa', 'resume_score']]
y = df['placed']
```

```python
X
```

```python
y
```

```python
# Now apply the model fro train the data
from sklearn.linear_model import Perceptron

P=Perceptron()
P.fit(X,y)
# Now we will predict the data
```

```python
P.coef_
```

```python
# There W1 and W2 are the weights of the model
# The model is trained and now we will predict the data
P.intercept_
# That is the bias of the model
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X.values, y.values, clf=P, legend=2)
plt.xlabel('cgpa')
plt.ylabel('resume_score')
plt.title('Perceptron')
plt.show()
```
