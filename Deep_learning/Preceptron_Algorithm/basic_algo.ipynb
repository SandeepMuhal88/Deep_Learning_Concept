```python
from sklearn.datasets import make_classification
import numpy as np
X,y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(X[:,0],X[:,1],c=y,edgecolor='k',s=50)
```

```python
def step(z):
    return 1 if z > 0 else 0
```

```python
def Precptron(X,y):
    X=np.insert(X,0,1,axis=1) # add bias term
    w=np.ones(X.shape[1]) # initialize weights
    lr=0.1 # learning rate
    epochs=1000 # number of epochs
    for i in range(epochs):
        j=np.random.randint(0,100) # random sample index
        y_hat=step(np.dot(X[j],w)) # predicted value
        w=w+lr*(y[j]-y_hat)*X[j] # update w
    
    return w[0],w[1:] # return bias and w
```

```python
intercept_,coef_ = Precptron(X,y)
```

```python
print(intercept_)
print(coef_)
```

```python
m=- coef_[0]/coef_[1] # slope of the line
b=intercept_/coef_[1] # y intercept
```
