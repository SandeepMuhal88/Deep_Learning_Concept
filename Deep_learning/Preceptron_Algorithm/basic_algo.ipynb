```python
def Preceptron(X,y):
    # Initialize weights and bias
    w = np.zeros(X.shape[1])
    b = 0
    learning_rate = 0.01
    epochs = 1000

    # Training loop
    for _ in range(epochs):
        for i in range(len(X)):
            # Calculate the linear output
            linear_output = np.dot(X[i], w) + b
            # Apply the activation function (step function)
            y_predicted = np.where(linear_output >= 0, 1, 0)
            # Update weights and bias if there is a misclassification
            update = learning_rate * (y[i] - y_predicted)
            w += update * X[i]
            b += update

    return w, b
```

```python
def pre(X,y):
    m=[]
    b=[]

    X=np.insert(X,0,1,axis=1) # add bias term
    w=np.ones(X.shape[1]) # initialize weights
    lr=0.1 # learning rate
    for i in range(1000):
        j=np.random.randint(0,100)
        y_hat=step(np.dot(X[j],w))
        w=w+lr*(y[j]-y_hat)*X[j]

        m.append(-(w[1]/w[2]))
        b.append(-(w[0]/w[2]))
    return m,b
```
