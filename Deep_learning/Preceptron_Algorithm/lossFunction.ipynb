```python
from sklearn.datasets import make_classification

import numpy as np

X,y=make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0,n_classes=2,n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=15)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter',s=100)
```

```python
def Perceptron(X,y):
    w1=w2=b=1
    lr=0.1
    for i in range(100):
        for j in range(X.shape[0]):
            if y[j]*(w1*X[j][0]+w2*X[j][1]+b)<=0:
                w1=w1+lr*y[j]*X[j][0]
                w2=w2+lr*y[j]*X[j][1]
                b=b+lr*y[j]
    return w1,w2,b
```

```python
b
```

```python
m=(w1/w2)
c=-(b/w2)
print(m,c)
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input+c
plt.figure(figsize=(10, 6))
plt.plot(x_input,y_input,color='red',linewidth=2)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
w1,w2,b=Perceptron(X,y)
def plot_decision_boundary(w1,w2,b):
    x1 = np.linspace(-3, 3, 100)
    x2 = -(w1*x1 + b)/w2
    plt.plot(x1, x2, color='red', label='Decision Boundary')
    plt.legend()
    plt.show()
```

```python
plot_decision_boundary(w1,w2,b)
```
